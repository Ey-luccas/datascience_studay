**Comandos de Visualização do pandas**

``display(variavel)``

**Descrição:** Exibe a estrutura do Dataframe importado

Uso: display(dataframe_variavel) Utilizado para visualização de colunas, index. Podendo ser personalizado dentro do parênteses.

Exemplo: display(dataframe['idade_pessoa'] <= 20)

``variavel.head()``

**Descrição:** Exibe as primeiras linhas do DataFrame importado.

Uso: variavel.head(n), onde n é o número de linhas a serem exibidas. Se omitido, exibe as primeiras 5 linhas por padrão.

Exemplo: df.head(10) mostra as primeiras 10 linhas do DataFrame df.

``variavel.tail()``

**Descrição:** Exibe as últimas linhas do DataFrame importado.

Uso: variavel.tail(n), onde n é o número de linhas a serem exibidas. Se omitido, exibe as últimas 5 linhas por padrão.

Exemplo: df.tail(10) mostra as últimas 10 linhas do DataFrame df.

``variavel.describe()``

**Descrição:** Fornece estatísticas resumidas do DataFrame, como contagem, média, desvio padrão, valores mínimos e máximos, e quartis.

Uso: variavel.describe()

Exemplo: df.describe() retorna as estatísticas descritivas do DataFrame df.

``variavel.info()``

**Descrição:** Fornece um resumo conciso do DataFrame, incluindo o número de entradas, o tipo de dados de cada coluna, e a memória utilizada.

Uso: variavel.info()

Exemplo: df.info() retorna um resumo informativo do DataFrame df.

``variavel.columns``

**Descrição:** Lista todas as colunas do DataFrame.

Uso: variavel.columns

Exemplo: df.columns retorna uma lista dos nomes das colunas no DataFrame df.

``variavel.shape``

**Descrição:** Retorna uma tupla representando a dimensão do DataFrame, ou seja, o número de linhas e colunas.

Uso: variavel.shape

Exemplo: df.shape retorna (número de linhas, número de colunas) do DataFrame df.

``variavel.isnull().sum()``

**Descrição:** Conta o número de valores nulos (missing values) em cada coluna do DataFrame.

Uso: variavel.isnull().sum()

Exemplo: df.isnull().sum() retorna uma série com a contagem de valores nulos em cada coluna do DataFrame df.

``variavel.groupby('coluna').mean()``

**Descrição:** Agrupa o DataFrame por uma coluna específica e calcula a média para cada grupo.

Uso: variavel.groupby('coluna').mean(), onde 'coluna' é a coluna pela qual você quer agrupar.

Exemplo: df.groupby('categoria').mean() agrupa o DataFrame df pela coluna 'categoria' e calcula a média de cada grupo.

``variavel.sort_values(by='coluna')``

**Descrição:** Ordena o DataFrame com base nos valores de uma coluna específica.

Uso: variavel.sort_values(by='coluna'), onde 'coluna' é a coluna pela qual você quer ordenar.

Exemplo: df.sort_values(by='idade') ordena o DataFrame df pela coluna 'idade'.
